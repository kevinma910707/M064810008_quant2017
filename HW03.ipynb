```python
import jieba
import jieba.analyse
```

```python
import requests
```

```python
url = 'https://news.cnyes.com/news/id/3966546?exp=a'
```

```python
page = requests.get(url)
page.encoding = "UTF-8"
```

```python
page.text
```

```python
from bs4 import BeautifulSoup
```

```python
#使用BeautifulSoup套件將文字檔存成html格式,並且搜尋文章主體
soup = BeautifulSoup(page.text,"lxml")
news_text = soup.select("div[itemprop=articleBody]")[0].text
print(news_text)
```

```python
#搜尋原文的摘要
news_summary = soup.p.string
print(news_summary)
```

```python
#嘗試抓取10個關鍵字
news_key = jieba.analyse.extract_tags(news_text,10,False)
```

```python
#用/分隔
print("/ ".join(news_key))
```

```python
#利用repr函數回傳表示每個關鍵字的機率
print (repr(jieba.analyse.extract_tags(news_text, topK=10, withWeight=True)))
```

```python
#利用repr函數回傳表示每個關鍵字的機率
print(repr(jieba.analyse.extract_tags(news_text, topK=20, withWeight=True)))#.decode('unicode-escape')
```

```python
#定義一個函數，找出關鍵字所在的段落
#enumerate函数可以同时列出數據和數據所在位子

def index_of_keyword(sentences, keyword):
    for i, s in enumerate(sentences):
        if keyword in s:
              return i
    return -1
```

```python
#將新聞\n刪除，並用句號分割數個段落
#計算有幾段


sentences = news_text.replace("\n","").split(u"。")
len(sentences)
```

```python
#在段落後面加入句號
for sentence in sentences:
    print(sentence + u"。")
```

```python
#利用的關鍵字索引,輸入sentences與頻蓋這兩個變數,尋找關鍵字的位置
index_of_keyword(sentences,u"蘋概")
```

```python
#顯示關鍵字的段落並加上句號
print(sentences[index_of_keyword(sentences,u"蘋概")]+u"。")
```

```python
##計算第一段有幾個字
len(sentences[0])
```

```python
s_index = []                                            #定義s_index為一個空陣列
for keyword in news_key[:4]:                            #利用迴圈找出前四個關鍵字所在的段落位置並放到s_index裡
    s_index.append(index_of_keyword(sentences,keyword))

s_set = set(s_index)                                    #利用set函數照順序排列，回傳到s_index裡
s_index = list(s_set)
```

```python
summary = ""                                            #定義summary為一個新的字串
for i in s_index:                                       #將剛剛所排好順序的段落位置轉成該段的文字回傳到summary裡
    summary += sentences[i]+u"。"
```

```python
print (summary) #將結果顯示出來
```
