# Q2. 寫一個可以產生金字塔的函式！

```python
def pyramid(n):
    for i in range(0,n): 
        print("" * (n-i-1) + "*" * (i))
        

pyramid(10)
```

```python
def printPyramid(level):
    for i in range(level):
        print (' ' * (level-i-1) + '*' * (2*i+1))
        
printPyramid(10)
```
