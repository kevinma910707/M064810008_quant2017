# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import requests

url= "https://finance.google.com/finance/historical"

params = {}
params['q'] = "DIS"
params['startdate'] = "2016-01-01"
params['output'] = 'csv'

result = requests.get(url, params=params)

print(result.text)
```
